# **Estudo de Caso Airbnb**

O Airbnb é um marketplace de aluguel de hospedagem. O seu formato de funcionamento é super simples: usuários que são donos de imóveis alugam o imóvel inteiro ou parte dele (um quarto, por exemplo), para um hóspede.

O Airbnb serve apenas como intermediador entre as partes. Seu papel é criar uma ponte entre os proprietários e os hóspedes, de forma que a relação entre eles seja clara e justa. O marketplace também realiza a mediação do pagamento, que é o momento no qual eles captam a comissão por cada negócio.

Note que o Airbnb não é proprietário de nenhum dos imóveis oferecidos na plataforma. Então, como o marketplace se assegura de que o negócio entre os envolvidos será feito de maneira justa?

Na verdade, existem diversas estratégias para reduzir a quantidade de problemas. Uma delas é o sistema de avaliação, que dá nota tanto para proprietários, como para os hóspedes. Assim, os usuários podem ter uma noção mais clara de quem está fazendo negócio.

Além disso, o próprio site possui ferramentas para garantir uma política de qualidade eficiente que evita, por exemplo, calotes, serviços mal feitos, imprecisões na contratação e outros problemas que podem vir a acontecer. Dessa forma, o Airbnb oferece segurança tanto para o anfitrião que aluga seu imóvel, quanto para o hóspede que usufrui por determinado período desse mesmo imóvel.

## **Outliers**

Na estatística, um outlier é um dado que se distancia radicalmente dos demais que compõem a amostra analisada. O termo é traduzido para “valor atípico” ou “valor aberrante” em português. Podemos dizer que é um “ponto fora da curva”, para usar uma expressão do dia a dia.

Como mencionado anteriormente, o foco da análise será dado em relação a variáveis price e minimum_nights.

Utilizando como base o histograma plotado destas variáveis conseguimos notar que elas não seguem uma distribuição e distorcem o histograma, portanto é um grande indicativo de presença de outliers.

Para nos auxiliar, vamos utilizar duas ferramentas para detecção de outliers:
* Resumo estatístico por meio do método describe(),
* Análise por meio do boxplot.

**Boxplot**

O Boxplot é uma ferramenta que nos auxilia, como o histograma, a avaliar o comportamento dos dados. Além de nos ajudar a identificar possíveis outliers.

O gráfico de boxplot é construído com base nas seguintes medidas de posição:

1º Quartil (Q1) - representa o valor que separa 25% dos valores inferiores e 75% dos valores superiores;
Mediana (Q2) - representa o valor central, ou seja, que separa 50% dos valores inferiores e 50% dos valores superiores. Esta medida também é chamada de 2º quartil;
3º Quartil (Q3) - representa o valor que separa 75% dos valores superiores e 25% dos valores inferiores.
Com estas informações é possível entender um pouco sobre a distribuição dos dados e sua dispersão.

Note que quanto maior for retângulo, formado por estas medidas, maior é sua dispersão. A partir da mediana é possível avaliar se a distribuição é simétrica ou assimétrica.

A figura abaixo é um exemplo de distribuição simétrica, ou seja, a mediana fica situada no meio do retângulo, separando as observações de forma igual. Portanto, podemos concluir que estes dados seguem uma distribuição normal.

im1.png

Como mencionado anteriormente, o boxplot é uma ferramenta que auxilia na detecção de outliers. Porém, precisamos de uma variável importante para delimitar os intervalos que podemos considerar como ponto fora do comportamento normal. Esta variável é chamada de Interquartil (IQR) que é gerada pela seguinte fórmula:

IQR = 3º Quartil - 1º Quartil
Porém, apenas com esta variável não é possível delimitar os intervalor de identificação de outlier. A literatura nos sugere a utilizar as seguintes medidas:

Limite Superior = 3º Quartil + 1,5 x IQR
Limite Inferior = 1º Quartil - 1,5 x IQR
Então, será considerado como outlier todo dado acima do limite superior e abaixo do limite inferior.

**Boxplot - variável price**

O próximo passo é retirar os outliers da variável price, porém devemos levar em consideração o fator da localidade, porque é comum que o preço do imóvel varie de acordo com o bairro.

Antes de iniciarmos este processo, devemos copiar o dataset original para uma outra variável para podermos alterar os dados a medida que retirarmos os outliers.

```python
#Copiando o dataset original
df_clean = df.copy()
```

Utilizaremos a cidade Hobart como exemplo de representação gráfica no boxplot.

```python
#Criando uma variável para armazenar os nomes das cidades
cidades_df = df['neighbourhood'].unique()

#Armazenando a quantidade de dados por cidade
contagem_cidade = df.groupby(['neighbourhood']).count()['id'].to_dict()

#Inicializando a figura do box plot
fig, ax = plt.subplots(figsize=(8, 4))

#Plotando o gráfico de boxplot
sns.set_theme(style="whitegrid")
sns.boxplot(x= df[df['neighbourhood'] == cidades_df[0]]['price'])

# remover grids e eixos
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# plotar título
plt.text(0, -0.6, 'Boxplot de ' + cidades_df[0],
         fontsize=16,
         color='#787878',
         weight='bold')

plt.show()
```

Para facilitar a retirada de outliers por cidade, executaremos uma linha de comando para retirar estes pontos de forma automática. Mas, vale ressaltar que nem todas as cidades possuem uma quantidade de dados razoável para aplicar o boxplot. Portanto, de forma arbitrária, será retirado outlier apenas dos bairros com mais de 4 eventos.

```python
#loop para retirar os outliers de forma automática
for cidade in cidades_df:

  #Verificação se a cidade tem 5 ou mais eventos
  if contagem_cidade[cidade ] >= 5:

    #Calculando as medidas necessárias para detectar outlier
    Q1 = df_clean[ df_clean['neighbourhood'] == cidade ]['price'].quantile(q = 0.25) #Definindo o primeiro quartil
    Q3 = df_clean[ df_clean['neighbourhood'] == cidade ]['price'].quantile(q = 0.75) #Definindo o segundo quartil
    IQR = Q3 - Q1 #Definindo o interquartil
    Limite_Superior = Q3 + IQR #Definindo o Limite Superior
    Limite_Inferior = Q1 - IQR #Definindo o Limite Inferior

    #Eliminando os outliers da amostra do bairro
    df_clean.drop(df_clean[(df_clean['neighbourhood'] == cidade) & (df_clean['price'] > Limite_Superior)].index, axis=0,inplace=True) #Retirando outliers superiores
    df_clean.drop(df_clean[(df_clean['neighbourhood'] == cidade) & (df_clean['price'] < Limite_Inferior)].index, axis=0,inplace=True) #Retirando outliers inferiores
```

Vamos reaplicar o resumo estatístico a partir do método describe() para avaliar se deixamos passar algo.

```python
df_clean[['price']].describe()
```

O describe()método retorna a descrição dos dados no DataFrame.

Se o DataFrame contiver dados numéricos, a descrição conterá essas informações para cada coluna:

# **Boxplot - variável minimum_nights**

Este passo visa retirar os outliers da variável minimum_nights. O processo será um pouco diferente da variável price, porque para este caso não há relação entre o mínimo de noites e a localidade.

Vamos plotar o boxplot desta variável.

```python
#Inicializando a figura do box plot
fig, ax = plt.subplots(figsize=(8, 4))

#Plotando o gráfico de boxplot
sns.set_theme(style="whitegrid")
sns.boxplot(x= df_clean['minimum_nights'])

# remover grids e eixos
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# plotar título
plt.text(0, -0.6, 'Boxplot - mínimo de noites',
         fontsize=16,
         color='#787878',
         weight='bold')

plt.show()
```

Vamos aplicar uma linha de comando para retirar os outliers dessa variável.

```python
#Calculando as medidas necessárias para detectar outlier
Q1 = df_clean['minimum_nights'].quantile(q = 0.25) #Definindo o primeiro quartil
Q3 = df_clean['minimum_nights'].quantile(q = 0.75) #Definindo o segundo quartil
IQR = Q3 - Q1 #Definindo o interquartil
Limite_Superior = Q3 + IQR #Definindo o Limite Superior
Limite_Inferior = Q1 - IQR #Definindo o Limite Inferior

#Eliminando os outliers da amostra do bairro
df_clean.drop(df_clean[df_clean['minimum_nights'] > Limite_Superior].index, axis=0,inplace=True) #Retirando outliers superiores
df_clean.drop(df_clean[df_clean['minimum_nights'] < Limite_Inferior].index, axis=0,inplace=True) #Retirando outliers inferiores
```

Retirado os outliers, vamos reaplicar o resumo estatístico por meio do método describe() para verificar se deixamos passar algo.

```python
#Reaplicando o método de resumo estatístico
df_clean[['minimum_nights']].describe()
```

De acordo com a tabela do resumo estatístico, não será necessário fazer nenhuma intervenção nos dados. Portanto, vamos para o próximo passo.
